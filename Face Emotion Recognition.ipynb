```python
import cv2
from deepface import DeepFace

# Load the pre-trained face detection model
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Capture video from webcam
cap = cv2.VideoCapture(1)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Try to open default webcam
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()

    # Check if the frame was successfully captured
    #if frame is None:
        #print("Failed to capture image")
        #break

    # Analyze the frame using DeepFace with enforce_detection set to False
    try:
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
    except ValueError as e:
        print("Face detection error:", e)
        result = {'dominant_emotion': 'Unknown'}

    # Convert the frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Prepare font for the text
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Display the dominant emotion
    cv2.putText(frame,
                result[0]['dominant_emotion'],  # Correct key for emotion
                (50, 50),
                font, 1,  # Adjusted font scale
                (0, 0, 255),
                2,
                cv2.LINE_4)

    # Show the video with detected faces and emotion
    cv2.imshow('Original Video', frame)

    # Exit if the 'q' key is pressed
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
from deepface import DeepFace

# Load the pre-trained face detection model
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Capture video from webcam
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)  # Try to open default webcam
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret, frame = cap.read()

    # Check if the frame was successfully captured
    if not ret or frame is None:
        print("Failed to capture image")
        break

    # Analyze the frame using DeepFace with enforce_detection set to False
    try:
        result = DeepFace.analyze(frame, actions=['emotion'], enforce_detection=False)
    except ValueError as e:
        print("Face detection error:", e)
        result = [{'dominant_emotion': 'Unknown'}]

    # Convert the frame to grayscale for face detection
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect faces in the frame
    faces = faceCascade.detectMultiScale(gray, 1.1, 4)

    # Draw rectangles around detected faces
    for (x, y, w, h) in faces:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Prepare font for the text
    font = cv2.FONT_HERSHEY_SIMPLEX

    # Display the dominant emotion
    cv2.putText(frame,
                result[0]['dominant_emotion'],  # Correct key for emotion
                (50, 50),
                font, 1,  # Adjusted font scale
                (0, 0, 255),
                2,
                cv2.LINE_4)

    # Show the video with detected faces and emotion
    cv2.imshow('Original Video', frame)

    # Exit if the 'q' key is pressed
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

# Release the video capture object and close all windows
cap.release()
cv2.destroyAllWindows()
```
